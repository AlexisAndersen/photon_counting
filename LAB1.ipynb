```python
import math
from scipy.interpolate import make_interp_spline

arrival_time = np.cumsum(filtered_dtH[0:2500])

fig, ax = plt.subplots(2,1, figsize=(8,8))
fig.tight_layout(pad=2.0)

ax[0].plot(range(len(arrival_time)), arrival_time,'go', markersize=2)
ax[0].set_xlabel('Event Number', fontsize = 12)
ax[0].set_ylabel('Time [ticks]', fontsize = 12)

counts_per_bin = np.histogram(arrival_time, bins = 1500)[0]
x=np.linspace(0,2e9,len(counts_per_bin))

#ax[1].plot(x, counts_per_bin)
#ax[1].set_xlabel('Time [ticks]', fontsize = 12)
#ax[1].set_ylabel('Counts per Bin', fontsize = 12)

z = 1
xpt = np.arange(0,7,1)
ypt = np.array([])
for i in range(0,7):
    y = 2300*(((z**xpt[i])/math.factorial(xpt[i]))*np.e**-z)
    ypt = np.append(ypt,y)
x_y_spline = make_interp_spline(xpt, ypt)
x_ = np.linspace(xpt.min(), xpt.max(), 1000)
y_ = x_y_spline(x_)

ax[1].hist(counts_per_bin, bins=7, edgecolor='green', linewidth=1.7, alpha=0.7, histtype="step")
ax[1].plot(x_, y_, color="y", drawstyle='steps-mid')
ax[1].set_xlabel('Counts per Bin', fontsize = 12)
ax[1].set_ylabel('Frequency', fontsize = 12)

plt.savefig("binhigh.png")
```
